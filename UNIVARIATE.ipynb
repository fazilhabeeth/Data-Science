```python
def QuanQual(dataset):
    Quan=[]#qualitative is object it means noun
    Qual=[]#quantitative is float it means numbers
    for column in dataset.columns:
        print(column)
        if(dataset[column].dtypes=='object'):
            Qual.append(column)#append=add; objects are added to qual list
        else:
            Quan.append(column)# float are added to the quan list
    return Quan,Qual
```

```python
def freqtable(columnName,dataset):
    frq=pd.DataFrame()
    frq["Unique_value"]=dataset[columnName].value_counts().index
    frq["frequency"]=dataset[columnName].value_counts().values
    frq["relative_freq"]=dataset[columnName].value_counts().values/len(dataset["ssc_p"])*100
    frq["cummulative_freq"]=frq["relative_freq"].cumsum()
    return frq
```

```python
freqtable("ssc_p",dataset) #our tabulation
```
