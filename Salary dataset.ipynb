```python
def freqtable(columnName,dataset):
    frq=pd.DataFrame()
    frq["Unique_value"]=dataset[columnName].value_counts().index
    frq["frequency"]=dataset[columnName].value_counts().values
    frq["relative_freq"]=dataset[columnName].value_counts().values/len(dataset["DATA_SCI"])*100
    frq["cummulative_freq"]=frq["relative_freq"].cumsum()
    return frq
```

```python
def QuanQual(dataset):
    Quan=[]
    Qual=[]
    for column in dataset.columns:
        print(column)
        if(dataset[column].dtypes=='object'):
            Qual.append(column)
        else:
            Quan.append(column)
    return Quan,Qual
```

```python
Univarite=pd.DataFrame(index=["Mean","Median","Mode","Min","25%","50%","75%","Max","IQR","1.5rule","Lesser","Greater","variance","std","kurtosis","skew"],columns=quan)
for column in quan:
    print(column)
    Univarite[column]["Mean"]=dataset[column].mean()
    Univarite[column]["Median"]=dataset[column].median() 
    Univarite[column]["Mode"]=dataset[column].mode()[0]
    Univarite[column]["Min"]=dataset[column].min(0)
    Univarite[column]["25%"]=np.percentile(dataset[column],25)
    Univarite[column]["50%"]=np.percentile(dataset[column],50)
    Univarite[column]["75%"]=np.percentile(dataset[column],75)
    Univarite[column]["Max"]=np.percentile(dataset[column],100)
    Univarite[column]["IQR"]=Univarite[column]["75%"]-Univarite[column]["25%"]
    Univarite[column]["1.5rule"]=1.5*Univarite[column]["IQR"]
    Univarite[column]["Lesser"]=Univarite[column]["75%"]-Univarite[column]["1.5rule"]
    Univarite[column]["Greater"]=Univarite[column]["25%"]+Univarite[column]["1.5rule"]
    Univarite[column]["variance"]=dataset[column].var()
    Univarite[column]["std"]=dataset[column].std()
    Univarite[column]["kurtosis"]=dataset[column].kurtosis()
    Univarite[column]["skew"]=dataset[column].skew()
```
